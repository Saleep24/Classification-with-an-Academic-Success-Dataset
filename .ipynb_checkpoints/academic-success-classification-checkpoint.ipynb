```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import keras_tuner as kt
import keras
from keras import backend as K
import tensorflow_decision_forests as tfdf
```

```python
pd.set_option('display.max_columns', 40)
```

```python
training_df = pd.read_csv('train.csv')
testing_df = pd.read_csv('test.csv')
```

```python
training_df.head()
```

```python
target_class = training_df['Target'].unique().tolist()
target_class
```

```python
def change_target_to_values(x):
    return target_class.index(x)

def change_values_to_target(x):
    return target_class[x]
```

```python
training_df['Target'] = training_df['Target'].apply(change_target_to_values)
```

```python
training_df.head()
```

```python
training_df.drop('id', axis=1).nunique()
```

```python
training_df.drop(['id'],axis=1).describe().drop('count').T
```

```python
training_df.info()
```

```python
misc_cols = ['id']
plt.figure(figsize=(20,20))
sns.heatmap(training_df.drop(misc_cols, axis=1).corr(), cmap='coolwarm', linewidth=0.1, robust=True, vmax=0.5, vmin=-0.5, center=0,)
```

```python
training_set = training_df.sample(frac=0.9, random_state = 420).drop('id', axis=1)
validation_set = training_df.drop(training_set.index).drop('id', axis=1)
```

```python
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(training_set, label='Target')
val_ds = tfdf.keras.pd_dataframe_to_tf_dataset(validation_set, label='Target')
```

```python
tuner = tfdf.tuner.RandomSearch(num_trials=50)

tuner.choice("min_examples", [2, 5, 7, 10, 20, 30])
tuner.choice("categorical_algorithm", ["CART", "RANDOM"])


local_search_space = tuner.choice("growing_strategy", ["LOCAL"])
local_search_space.choice("max_depth", [3, 4, 5, 6, 8])

global_search_space = tuner.choice("growing_strategy", ["BEST_FIRST_GLOBAL"], merge=True)
global_search_space.choice("max_num_nodes", [16, 32, 64, 128, 256])

tuner.choice("use_hessian_gain", [True, False])
tuner.choice("shrinkage", [0.02, 0.05, 0.10, 0.15])
tuner.choice("num_candidate_attributes_ratio", [0.2, 0.5, 0.9, 1.0])
```

```python
model= tfdf.keras.GradientBoostedTreesModel(tuner=tuner)
model.fit(train_ds)
```

```python
model.compile(metrics=["accuracy"])
evaluation = model.evaluate(val_ds, return_dict=True)

for name, value in evaluation.items():
  print(f"{name}: {value:.4f}")
```

```python
submission_ds = tfdf.keras.pd_dataframe_to_tf_dataset(testing_df.drop('id', axis=1))
```

```python
preds = model.predict(submission_ds)
```

```python
testing_df
```

```python
testing_df['Target'] = [np.argmax(i) for i in preds]
```

```python
testing_df['Target'] = testing_df['Target'].apply(change_values_to_target)
```

```python
testing_df
```

```python
submission = testing_df[['id','Target']]
```

```python
submission.to_csv('kapil_submission_v9.csv', index=False)
```
